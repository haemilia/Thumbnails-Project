```python
df = pd.read_csv('2022.csv', index_col=0)
```

```python
#get image from url 
def url_to_img(url_lst):
    for i in range(len(url_lst)):
        img_url = url_lst[i]
        file_name = f"thumbnail_img/{i}.png"
        urllib.request.urlretrieve(img_url, file_name)
        print(f"{file_name}.png saved")    
```

```python
url_to_img(df['ThumbnailUrl'])
```
