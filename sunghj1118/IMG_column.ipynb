```python
# path ì„¤ì •
DATA_PATH = '/Users/sungh/OneDrive/Desktop/THUMBNAILS/test/'
```

```python
def getImages(df, IDX) :

    # input : dataset to get channel id and video id
    # output : dataset with image name

    # create columns
    df['Image'] = 0

    for i in range(len(df)):
        img_id = df['Unnamed: 0'][i]
        img_url = df['HDThumbnailUrl'][i]

        file_name = f"{IDX}_{img_id}.png"
        df['Image'][i] = file_name
    
    return df
```

```python
fullData = pd.DataFrame({
    'ChannelName' : [],
    'VideoIndex': [],
    'VideoID' : [],
    'Title': [],
    'Image' : []
})
```

```python
for idx in range(len(YOUTUBE_LIST)):
    NAME = YOUTUBE_LIST[idx]
    data = pd.read_csv(DATA_PATH + NAME + '.csv')
    dataLabeled = getImages(data, idx)
    
    mainData = pd.DataFrame({
        'ChannelName' : idx,
        'VideoIndex': dataLabeled['Unnamed: 0'],
        'VideoID' : dataLabeled['VideoId'],
        'Title': dataLabeled['Title'],
        'Image': dataLabeled['Image']
    })
    
    fullData = pd.concat([fullData, mainData])
```
